***Book Dataset***

```python
book = pd.read_csv("book.csv")
book.head()
```

### Apriori Algorithm 

```python
frequent_itemsets_1 = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets_1
```

### Rule 1 ###

```python
rule_1 = association_rules(frequent_itemsets_1, metric="lift", min_threshold=0.7)
rule_1
```

```python
rule_1.sort_values('lift',ascending = False)[0:20]
```

```python
rule_1[rule_1.lift>1]
```

***Movies Dataset***

```python
Movies = pd.read_csv("Movies.csv")
Movies.head()
```

### Pre-Processing
As the data is not in transaction formation 
We are using transaction Encoder

```python
df=pd.get_dummies(Movies)
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
