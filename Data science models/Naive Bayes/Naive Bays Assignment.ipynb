```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB as GB
import numpy as np
```

```python
data_train = pd.read_csv("SalaryData_Train.csv")
```

```python
data_test = pd.read_csv("SalaryData_Test.csv")
```

```python
data_train 
```

```python
data_test
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]   
for x in string_columns:
    data_test[x] = data_test[x].astype(str)
    
for i in string_columns:
    number = preprocessing.LabelEncoder()
    data_test[i] = number.fit_transform(data_test[i])
```

```python
data_test
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]   
for x in string_columns:
    data_train [x] = data_train[x].astype(str)
    
for i in string_columns:
    number = preprocessing.LabelEncoder()
    data_train[i] = number.fit_transform(data_train[i])
```

```python
data_train 
```

```python
array1 = data_train.values
X1 = array1[:, 0:13]
Y1 = array1[:, 13]

array2 = data_test.values
X2 = array2[:, 0:13]
Y2 = array2[:, 13]
```

```python
classifier_gb = GB()
classifier_gb.fit(X1,Y1) 
train_pred_g = classifier_gb.predict(X1)
accuracy_train_g = np.mean(train_pred_g==Y1) 
test_pred_g = classifier_gb.predict(X2)
accuracy_test_g = np.mean(test_pred_g==Y2)
```
