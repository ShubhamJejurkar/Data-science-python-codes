# Imdb Reviews

```python
import requests
from bs4 import BeautifulSoup
```

```python
#url = "http://www.imdb.com/chart/top"
url = "http://www.imdb.com/title/tt3521164/reviews?ref_=tt_urv"
resp = requests.get(url)
print(resp.status_code)
#print(resp.content)
soup = BeautifulSoup(resp.text,"html.parser")
#soup.prettify()
```

```python
reviews = soup.findAll("div", attrs={"class","text"})
print(type(reviews))
for i in range(len(reviews)):
    reviews[i] = reviews[i].text
    #print(reviews[i])
    #print("____________________________________")
    
```

```python
# Creating a data frame 
import pandas as pd
movie_reviews = pd.DataFrame(columns = ["reviews"])
movie_reviews["reviews"] = reviews
movie_reviews
```

## Top Movies

```python
url = "http://www.imdb.com/chart/top"
#url = "http://www.imdb.com/title/tt3521164/reviews?ref_=tt_urv"
resp = requests.get(url)
#print(resp.status_code)
#print(resp.content)
soup = BeautifulSoup(resp.text,"html.parser")
#soup.prettify()
```

```python
llist = soup.find("tbody", attrs= {"class","lister-list"})
top_list=[]
trs = llist.findAll("tr")
for tr in trs:
    #print(tr.find("td",{"class":"titleColumn"}).text)
    item = tr.find("td",attrs={"class","titleColumn"}).text
    top_list.append(item.replace('\n',''))
top_list    
```
