```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import StandardScaler

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

### Salary Dataset

```python
data_train = pd.read_csv("/content/SalaryData_Train.csv")
```

```python
data_test = pd.read_csv("/content/SalaryData_Test.csv")
```

```python
data_train 
```

```python
data_test
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]   
for x in string_columns:
    data_test[x] = data_test[x].astype(str)
    
for i in string_columns:
    number = preprocessing.LabelEncoder()
    data_test[i] = number.fit_transform(data_test[i])
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]   
for x in string_columns:
    data_train [x] = data_train[x].astype(str)
    
for i in string_columns:
    number = preprocessing.LabelEncoder()
    data_train[i] = number.fit_transform(data_train[i])
```

```python
array1 = data_train.values
X1 = array1[0:1500, 0:13]
Y1 = array1[0:1500, 13]

array2 = data_test.values
X2 = array2[:, 0:13]
Y2 = array2[:, 13]
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X1,Y1)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,48,47,51,52],'C':[15,16,17,18] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X1,Y1)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
array3 = data_train.values
X3 = array1[:, 0:13]
Y3 = array1[:, 13]
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(X3 , Y3)
y_pred = clf.predict(X2)
acc = accuracy_score(Y2, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(Y2, y_pred)
```

### Forest fires Dataset

```python
Data = pd.read_csv("/content/forestfires.csv")
```

```python
Data
```

```python
string_columns = ["month","day","size_category"]   

for x in string_columns:
    Data[x] = Data[x].astype(str)
    
for i in string_columns:
    number = preprocessing.LabelEncoder()
    Data[i] = number.fit_transform(Data[i])
```

```python
Data
```

```python
array = Data.values
X = array[:,0:19]
Y = array[:,19]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[0.0001,0.1,0.00001],'C':[15,16,17,18] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 0.0001)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
