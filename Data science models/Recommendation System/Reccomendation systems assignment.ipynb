```python
import pandas as pd
import numpy as np
```

```python
book_df = pd.read_csv('book.csv')
```

```python
book_df
```

```python
book_df = book_df.rename({'Book.Title': 'Book','Book.Rating':'Rating','User.ID':'UserID'}, axis=1)
```

```python
book_df
```

```python
book_df=book_df.drop('Unnamed: 0',axis=1)
```

```python
book_df
```

```python
book_df.shape
```

```python
data_cleaned=book_df.drop_duplicates(subset = ["UserID"])
```

```python
user_book_df = data_cleaned.pivot(index='UserID',columns='Book', values='Rating').reset_index(drop=True)
```

```python
user_book_df 
```

```python
user_book_df.index = book_df.UserID.unique()
```

```python
user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = book_df.UserID.unique()
user_sim_df.columns = book_df.UserID.unique()
```
